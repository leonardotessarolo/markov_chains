```python
import numpy as np
from hmm import HMM

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
N=50
y=np.random.binomial(
    n=1,
    p=0.5,
    size=(N,)
)
hmm = HMM(
    n_hidden_states=2,
    observed_vocabulary=[0,1]
)

hmm.fit(
    y=y,
    thresh=0.0001
)
```

```python
np.sum(hmm.gamma[:,0])/N, np.sum(hmm.gamma[:,1])/N
```

```python
fig = plt.figure(figsize=(20,7))
plt.plot(
    range(1,len(hmm.prob_evolution)+1),
    np.log10(hmm.prob_evolution)
)
```
